```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 경고 무시 설정
import warnings
warnings.filterwarnings(action='ignore')

# 한글 깨짐 방지
plt.rcParams['font.family'] = 'Malgun Gothic'
# 마이너스 부호 깨짐 방지
plt.rcParams['axes.unicode_minus'] = False
```

```python
df0 = pd.read_csv('./spieces_data.csv')
df0.columns = df0.columns.str.strip()
df0
```

```python
# 1. '대지위치' 컬럼 삭제
df00 = df0.drop(columns=['대지위치'])

# 2. 법정동코드 + 번 + 지 를 합쳐서 'code' 컬럼 생성
# 문자열 연산을 통해 하나로 합칩니다.
df00['code'] = df0['법정동코드'].astype(str) + '-' + \
                     df0['번'].astype(str) + '-' + \
                     df0['지'].astype(str)
    
# 3. 합치기에 사용된 기존 컬럼 삭제 (선택 사항)
df00= df00.drop(columns=['법정동코드', '번', '지'])

# 4. 필요한 컬럼만 선택
df00 = df00[['code', '주용도코드명']]
df00
```

```python
df00.to_csv('species.csv', index=False, encoding='utf-8-sig')
```
