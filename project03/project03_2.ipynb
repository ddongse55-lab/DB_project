## 2020년-2023년 건물용도별(주택/산업) 전력사용량 분석

### 가정용 전력사용량

```python
# 파일을 불러올 때 상단 5줄(0~4번)이 모두 제목이라고 알려줘요.
df_home = pd.read_csv('가정용_전력_사용량_2020_2023.csv')
df_home.head()
```

```python
# '원데이터'라고 써진 기둥(열)만 쏙 뽑아낼 거예요.
# 월별(1) 컬럼과 각 연도의 '원데이터' 컬럼만 선택합니다.
cols_to_keep = [col for col in df_home.columns if '원데이터' in col or '월별' in col[0]]
df_home_clean = df_home[cols_to_keep]

# 지우고 나서 순서를 다시 0, 1, 2... 순서대로 정리해요.
df_home_clean = df_home_clean.reset_index(drop=True)

# 제목을 알기 쉽게 바꿔요. 첫 번째는 '월', 나머지는 연도만 남겨요.
df_home_clean.columns = ['월', '2020년', '2021년', '2022년', '2023년']
df_home_clean
```

```python
df_home_clean.to_csv('home_data_2020_2023_v2.csv', index=False, encoding='utf-8')
```
