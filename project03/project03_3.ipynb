#### 제조업

```python
# 1. 파일 불러오기
data1_summary = pd.read_csv('./data1_summary.csv')

# 2. 데이터가 잘 왔는지 확인하기
data1_summary
```

```python
# 연도별로 막대그래프 그리기
colors = plt.cm.Set3(np.linspace(0, 1, 4))
plt.figure(figsize=(10, 6))
plt.bar(['2020년', '2021년', '2022년', '2023년'], 
        [data1_summary['2020년_제조업_합계'][0], data1_summary['2021년_제조업_합계'][0], 
         data1_summary['2022년_제조업_합계'][0], data1_summary['2023년_제조업_합계'][0]],
        color=colors)

plt.title('연도별 전력 사용량 합계(산업용-제조업)')
plt.xlabel('연도')
plt.ylabel('사용량(MWh)')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. 데이터 불러오기
df = pd.read_csv('data1_summary.csv')

# 2. 데이터 정리 (컬럼명에서 연도만 추출하여 깔끔하게 정리)
# 컬럼: 2020년_제조업_합계 -> 2020년
years = [col.split('_')[0] for col in df.columns]
values = df.iloc[0].values  # 첫 번째 행의 수치 데이터들

# 3. 그래프 설정
plt.rcParams['font.family'] = 'Malgun Gothic'  # 한글 폰트 설정
plt.rcParams['axes.unicode_minus'] = False     # 마이너스 기호 깨짐 방지

plt.figure(figsize=(10, 6))
colors = ['#4A90E2', '#50E3C2', '#F5A623', '#D0021B'] # 연도별 구분 색상

# 막대 그래프 그리기 (결과를 'bars' 변수에 담아야 수치 표시가 가능합니다)
bars = plt.bar(years, values, color=colors, alpha=0.8, edgecolor='black', linewidth=1)

# 4. 막대 상단에 데이터 수치 적기
for bar in bars:
    height = bar.get_height()
    # 수치 표시: 천 단위 콤마(,) 추가 / 위치는 막대 높이보다 2% 높게 설정
    plt.text(bar.get_x() + bar.get_width()/2., height + (max(values) * 0.02),
             f'{int(height):,}', 
             ha='center', va='bottom', fontsize=11, fontweight='bold', color='black')

# 5. 그래프 꾸미기
plt.title('연도별 제조업 전력 사용량 합계 (2020-2023)', fontsize=16, pad=20)
plt.xlabel('연도', fontsize=12)
plt.ylabel('전력 사용량 (MWh)', fontsize=12)

# Y축 범위를 수치 표시 공간 확보를 위해 15% 정도 높게 설정
plt.ylim(0, max(values) * 1.15)

plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()

plt.show()
```

```python
# 연도별로 막대그래프 그리기
colors = ['blue', 'orange', 'gold', 'green']
plt.figure(figsize=(10, 6))
plt.bar(['2020년', '2021년', '2022년', '2023년'], 
        [data1_summary['2020년_제조업_합계'][0], data1_summary['2021년_제조업_합계'][0], 
         data1_summary['2022년_제조업_합계'][0], data1_summary['2023년_제조업_합계'][0]],
        color=colors)

plt.title('연도별 전력 사용량 합계(산업용-제조업)')
plt.xlabel('연도')
plt.ylabel('사용량(MWh)')
plt.show()
```

### 산업용 전력사용양 연도별 누적 막대 그래프

```python
# 1. 각각 저장했던 요약 파일들 불러오기
df1 = pd.read_csv('data1_summary.csv') # 제조업
df2 = pd.read_csv('data2_summary.csv') # 광업
df3 = pd.read_csv('data3_summary.csv') # 농림수산업

# 2. 그래프를 그리기 쉽게 데이터 정리하기
years = ['2020년', '2021년', '2022년', '2023년']

# 각 파일에서 숫자 데이터만 쏙쏙 뽑아서 리스트로 만들어요
mfg = [df1.iloc[0, 0], df1.iloc[0, 1], df1.iloc[0, 2], df1.iloc[0, 3]] # 제조업
mining = [df2.iloc[0, 0], df2.iloc[0, 1], df2.iloc[0, 2], df2.iloc[0, 3]] # 광업
agri = [df3.iloc[0, 0], df3.iloc[0, 1], df3.iloc[0, 2], df3.iloc[0, 3]] # 농림수산업

# 3. 누적 막대 그래프 그리기
plt.figure(figsize=(10, 7))

# 맨 아래에 제조업을 깔아요
plt.bar(years, mfg, label='제조업', color='royalblue')

# 그 위에 광업을 쌓아요 (bottom 파라미터가 중요!)
plt.bar(years, mining, bottom=mfg, label='광업', color='gold')

# 마지막으로 그 위에 농림수산업을 쌓아요
# 밑에 깔린 게 제조업+광업이니까 두 개를 더한 값을 bottom으로 줘요
bottom_for_agri = [mfg[i] + mining[i] for i in range(len(mfg))]
plt.bar(years, agri, bottom=bottom_for_agri, label='농림수산업', color='limegreen')

# # 수치 추가를 위한 로직
# for i in range(len(years)):
#     # 1. 제조업 (맨 아래)
#     plt.text(i, mfg[i]/2, f'{int(mfg[i]):,}', ha='center', va='center', color='white', fontweight='bold')
    
#     # 2. 광업 (제조업 위)
#     plt.text(i, mfg[i] + mining[i]/2, f'{int(mining[i]):,}', ha='center', va='center', color='black', fontweight='bold')
    
#     # 3. 농림수산업 (맨 위)
#     current_bottom = mfg[i] + mining[i]
#     plt.text(i, current_bottom + agri[i]/2, f'{int(agri[i]):,}', ha='center', va='center', color='black', fontweight='bold')
    
#     # 4. 총합 (막대 맨 꼭대기)
#     total = mfg[i] + mining[i] + agri[i]
#     plt.text(i, total + (total * 0.01), f'{int(total):,}', ha='center', va='bottom', color='black', fontsize=10)

# 4. 그래프 꾸미기
plt.title('연도별 산업용 전력 사용량 구성 (누적 막대)', fontsize=16)
plt.ylabel('사용량 (MWh)', fontsize=12)
plt.legend(title='용도별')
plt.grid(axis='y', linestyle='--', alpha=0.5)

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. 데이터 불러오기
df1 = pd.read_csv('data1_summary.csv') # 제조업
df2 = pd.read_csv('data2_summary.csv') # 광업
df3 = pd.read_csv('data3_summary.csv') # 농림수산업

# 2. 데이터 추출 및 정리
years = ['2020년', '2021년', '2022년', '2023년']
mfg = df1.iloc[0].values     # 제조업
mining = df2.iloc[0].values  # 광업
agri = df3.iloc[0].values    # 농림수산업

# 3. 그래프 설정
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False
sns.set_style("whitegrid")

plt.figure(figsize=(12, 8))
colors = sns.color_palette("husl", 3) # 용도별 구분을 위한 3가지 색상

# 4. 누적 막대 그래프 그리기
# 아래에서부터 제조업 -> 광업 -> 농림수산업 순으로 쌓음
p1 = plt.bar(years, mfg, label='제조업', color=colors[0], alpha=0.85, edgecolor='white')
p2 = plt.bar(years, mining, bottom=mfg, label='광업', color=colors[1], alpha=0.85, edgecolor='white')

bottom_for_agri = mfg + mining
p3 = plt.bar(years, agri, bottom=bottom_for_agri, label='농림수산업', color=colors[2], alpha=0.85, edgecolor='white')

# 5. 막대 상단에 총합 수치 기입
total_values = mfg + mining + agri
for i, total in enumerate(total_values):
    plt.text(i, total + (max(total_values) * 0.01), 
             f'{int(total):,}', 
             ha='center', va='bottom', fontsize=12, fontweight='bold', color='black')

# 6. 그래프 디자인 마무리
plt.title('연도별 산업용 전력사용량 구성 및 총합 (2020-2023)', fontsize=18, pad=25, fontweight='bold')
plt.ylabel('전력 사용량 (MWh)', fontsize=13)
plt.legend(title='용도별 구분', bbox_to_anchor=(1.05, 1), loc='upper left')

# 수치가 잘리지 않도록 y축 범위 조정
plt.ylim(0, max(total_values) * 1.15)

plt.tight_layout()
plt.show()
```
