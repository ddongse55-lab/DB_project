```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 경고 무시 설정
import warnings
warnings.filterwarnings(action='ignore')

# 한글 깨짐 방지
plt.rcParams['font.family'] = 'Malgun Gothic'
# 마이너스 부호 깨짐 방지
plt.rcParams['axes.unicode_minus'] = False
```

## 데이터 전처리(연도별 전기사용량)

### 2020년도 데이터

```python
df1 = pd.read_csv('./e_2020.csv')
# 컬럼명 앞뒤 공백 제거
df1.columns = df1.columns.str.strip()
df1
```

```python
# 1. '대지위치' 컬럼 삭제
df01 = df1.drop(columns=['대지위치'])

# 2. 법정동코드 + 번 + 지 를 합쳐서 'code' 컬럼 생성
# 문자열 연산을 통해 하나로 합칩니다.
df01['code'] = df1['법정동코드'].astype(str) + '-' + \
                     df1['번'].astype(str) + '-' + \
                     df1['지'].astype(str)
    
# 3. 합치기에 사용된 기존 컬럼 삭제 (선택 사항)
df01 = df01.drop(columns=['법정동코드', '번', '지'])

# 4. 사용년월에서 month 추출 및 사용년월(year 포함) 삭제
# 문자열로 변환 후 뒤에서 두 자리만 가져옵니다.
df01['year'] = df1['사용년월'].astype(str).str[:4]
df01['month'] = df1['사용년월'].astype(str).str[-2:]
df01 = df01.drop(columns=['사용년월'])

# 5. 사용량(KWh)을 계산을 위해 숫자형으로 변환
df01['e_amount'] = pd.to_numeric(df1['사용량(KWh)'], errors='coerce').fillna(0)

# 6. 필요한 컬럼만 선택
df01 = df01[['code', 'year','month', 'e_amount']]
df01
```

```python
df01.to_csv('e_code_2020.csv', index=False, encoding='utf-8-sig')
```
