```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 경고 무시 설정
import warnings
warnings.filterwarnings(action='ignore')

# 한글 깨짐 방지
plt.rcParams['font.family'] = 'Malgun Gothic'
# 마이너스 부호 깨짐 방지
plt.rcParams['axes.unicode_minus'] = False
```

## 2023년도 월별 전력량

```python
# 1. 파일 불러오기
df = pd.read_csv('./2023_usage.csv')

# 2. 데이터가 잘 왔는지 확인하기
df
```

```python
# 1. 월 순서대로 정렬하기 (중요!)
# '월'에서 숫만 뽑아서 순서를 정해줄 거예요
df['월_숫자'] = df['month_name'].str.replace('월', '').astype(int)
df = df.sort_values('월_숫자').reset_index(drop=True)
df
```

```python
# 1. 그래프 그리기 시작!
plt.figure(figsize=(12, 6))

# 가정용 (따뜻한 주황색)
plt.plot(df['month_name'], df['2023_가정용_전력사용량'], 
         marker='o', label='가정용', color='orange', linewidth=2)

# 제조업 (차갑고 신뢰감 있는 파란색)
plt.plot(df['month_name'], df['2023_제조업_전력사용량'], 
         marker='s', label='제조업', color='royalblue', linewidth=2)

# 농림수산업 (자연의 초록색)
plt.plot(df['month_name'], df['2023_농림수산업_전력사용량'], 
         marker='^', label='농림수산업', color='green', linewidth=2)

# 광업 (차분한 회색)
plt.plot(df['month_name'], df['2023_광업_전력사용량'], 
         marker='d', label='광업', color='gray', linewidth=2)

# 2. 그래프 꾸미기
plt.title('2023년 용도별 월별 전력사용량 추이', fontsize=15)
plt.xlabel('월', fontsize=12)
plt.ylabel('사용량 (MWh)', fontsize=12)
plt.legend() # 범례(이름표) 보여주기
plt.grid(True, linestyle='--', alpha=0.7) # 뒤에 격자 무늬 넣기

plt.show()
```

```python
# 1. 데이터 불러오기 및 월 순서 정렬
df_2023 = pd.read_csv('2023_usage.csv')
df_2023['월_숫자'] = df_2023['month_name'].str.replace('월', '').astype(int)
df_2023 = df_2023.sort_values('월_숫자')

# 2. 2행 2열(총 4개)의 도화지 준비하기
fig, axes = plt.subplots(2, 2, figsize=(15, 10))
fig.suptitle('2023년 용도별 월별 전력사용량', fontsize=20)

# --- 첫 번째: 가정용 (0행 0열) ---
axes[0, 0].plot(df_2023['month_name'], df_2023['2023_가정용_전력사용량'], marker='o', color='orange')
axes[0, 0].set_title('가정용')
axes[0, 0].grid(True, alpha=0.3)

# --- 두 번째: 제조업 (0행 1열) ---
axes[0, 1].plot(df_2023['month_name'], df_2023['2023_제조업_전력사용량'], marker='s', color='royalblue')
axes[0, 1].set_title('제조업')
axes[0, 1].grid(True, alpha=0.3)

# --- 세 번째: 광업 (1행 0열) ---
axes[1, 0].plot(df_2023['month_name'], df_2023['2023_광업_전력사용량'], marker='d', color='gray')
axes[1, 0].set_title('광업')
axes[1, 0].grid(True, alpha=0.3)

# --- 네 번째: 농림수산업 (1행 1열) ---
axes[1, 1].plot(df_2023['month_name'], df_2023['2023_농림수산업_전력사용량'], marker='^', color='green')
axes[1, 1].set_title('농림수산업')
axes[1, 1].grid(True, alpha=0.3)

# 3. 그래프 간격 조절하고 보여주기
plt.tight_layout(rect=[0, 0.03, 1, 0.95]) # 제목과 그래프가 겹치지 않게 조절
plt.show()
```
