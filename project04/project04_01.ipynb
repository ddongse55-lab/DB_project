```python
df = df03.iloc[8]
df
```

```python
df03_usage = pd.read_csv('./영양소별 섭취량 - 성별(남).csv')
df03_usage
```

```python
import pandas as pd

# 1. 인덱스 이름 통일 (계산 및 병합 오류 방지)
# df03_usage의 인덱스를 df(권장량)의 인덱스와 완벽히 일치시킵니다.
df03_usage.index = df.index

# 2. 두 테이블 합치기 (pd.concat 사용)
# keys를 사용하면 '실제'와 '권장'을 구분하는 큰 카테고리가 생겨서 관리가 쉽습니다.
combined_df = pd.concat([df03_usage, df], axis=1, keys=['실제섭취량', '권장섭취량'])
combined_df
```

```python
import pandas as pd

# 1. 불필요한 이중 인덱스(Level) 제거 및 열 정리
# 만약 멀티인덱스로 되어 있다면 단일 레벨로 내립니다.
combined_df.columns = combined_df.columns.get_level_values(-1)

# 2. '영양소' 컬럼이 데이터 안에 포함되어 있다면 삭제 (이미 인덱스에 있으므로)
if '영양소' in combined_df.columns:
    combined_df = combined_df.drop(columns=['영양소'])

# 3. 특정 컬럼명(인덱스 이름) 변경
# '평균' -> '실제섭취량', '8' -> '권장섭취량'
# (나머지 연령대 컬럼은 유지하거나 필요한 것만 필터링합니다)
combined_df = combined_df.rename(columns={'평균': '실제섭취량', '8': '권장섭취량'})

# 5. 결과 확인
print("--- 정리된 최종 데이터프레임 ---")
combined_df
```

```python
import pandas as pd

# 방법 1: 숫자 8과 문자열 '8' 모두 대응하기
combined_df = combined_df.rename(columns={8: '권장섭취량', '8': '권장섭취량'})

# 방법 2: 열 위치(Index)로 강제 변경 (가장 확실함)
# 만약 '8'이 세 번째(인덱스 2번) 열이라면 아래와 같이 수정 가능합니다.
# columns 리스트를 직접 확인하고 해당 위치의 이름을 바꿉니다.
cols = list(combined_df.columns)
if '8' in cols:
    cols[cols.index('8')] = '권장섭취량'
elif 8 in cols:
    cols[cols.index(8)] = '권장섭취량'
combined_df.columns = cols

# 방법 3: 섭취 비율 계산 시 바로 적용하기
# 이름이 안 바뀌더라도 아래처럼 직접 계산에 활용할 수 있습니다.
combined_df['섭취비율(%)'] = (combined_df['실제섭취량'] / combined_df.iloc[:, 1]) * 100 # 2번째 열이 권장량일 경우
combined_df
```

```python
import matplotlib.pyplot as plt

# 1. 폰트 설정 (한글 깨짐 방지)
plt.rcParams['font.family'] = 'Malgun Gothic' # 윈도우 기준
plt.rcParams['axes.unicode_minus'] = False

# 2. 모든 영양소에 대해 막대그래프 그리기
plt.figure(figsize=(15, 8))

# x축은 영양소 인덱스, y축은 계산하신 '섭취비율(%)'
# 만약 '섭취비율(%)' 열이 아직 없다면 아래 주석을 해제하고 실행하세요.
# df_final['섭취비율(%)'] = (df_final['실제섭취량'] / df_final.iloc[:, 1]) * 100

bars = plt.bar(combined_df.index, combined_df['섭취비율(%)'], color='skyblue', edgecolor='navy', alpha=0.7)

# 3. 100% 위치에 빨간색 기준선 그리기
plt.axhline(y=100, color='red', linestyle='--', linewidth=2, label='권장 섭취량 기준 (100%)')

# 4. 그래프 꾸미기
plt.title('영양소별 권장량 대비 실제 섭취 비율 - 남성 (%)', fontsize=18, pad=20)
plt.ylabel('섭취 비율 (%)', fontsize=12)
plt.xlabel('영양소', fontsize=12)
plt.xticks(rotation=45, ha='right') # 영양소 이름이 길 경우를 대비해 회전
plt.grid(axis='y', linestyle=':', alpha=0.6)
plt.legend()

# 5. 막대 위에 수치 표시 (가독성 향상)
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height + 1,
             f'{height:.1f}%', ha='center', va='bottom', fontsize=9)

plt.tight_layout()
plt.show()
```

```python
df04_usage = pd.read_csv('./영양소별 섭취량 - 성별(여).csv')
df04_usage
```

```python
import pandas as pd

# 1. 인덱스 이름 통일 (계산 및 병합 오류 방지)
# df03_usage의 인덱스를 df(권장량)의 인덱스와 완벽히 일치시킵니다.
df04_usage.index = df2.index

# 2. 두 테이블 합치기 (pd.concat 사용)
# keys를 사용하면 '실제'와 '권장'을 구분하는 큰 카테고리가 생겨서 관리가 쉽습니다.
combined_df = pd.concat([df04_usage, df2], axis=1, keys=['실제섭취량', '권장섭취량'])
combined_df
```

```python
import pandas as pd

# 1. 불필요한 이중 인덱스(Level) 제거 및 열 정리
# 만약 멀티인덱스로 되어 있다면 단일 레벨로 내립니다.
combined_df.columns = combined_df.columns.get_level_values(-1)

# 2. '영양소' 컬럼이 데이터 안에 포함되어 있다면 삭제 (이미 인덱스에 있으므로)
if '영양소' in combined_df.columns:
    combined_df = combined_df.drop(columns=['영양소'])

# 3. 특정 컬럼명(인덱스 이름) 변경
# '평균' -> '실제섭취량', '8' -> '권장섭취량'
# (나머지 연령대 컬럼은 유지하거나 필요한 것만 필터링합니다)
combined_df = combined_df.rename(columns={'평균': '실제섭취량', '8': '권장섭취량'})

# 5. 결과 확인
print("--- 정리된 최종 데이터프레임 ---")
combined_df
```

```python
# 2. 데이터 전처리 (숫자형 변환 및 결측치 제거)
# errors='coerce'를 사용해 숫자가 아닌 데이터는 NaN으로 만든 뒤 0으로 채웁니다.
combined_df = combined_df.apply(pd.to_numeric, errors='coerce').fillna(0)

# 3. 열 이름 강제 재설정 (오류 방지)
# 첫 번째 열을 '실제섭취량', 두 번째 열을 '권장섭취량'으로 강제 지정합니다.
combined_df.columns = ['실제섭취량', '권장섭취량']
combined_df
```

```python
# 4. 섭취비율(%) 계산
combined_df['섭취비율(%)'] = (combined_df['실제섭취량'] / combined_df['권장섭취량']) * 100
# 0으로 나눠서 발생하는 무한대(inf) 값을 0으로 처리
combined_df['섭취비율(%)'] = combined_df['섭취비율(%)'].replace([float('inf'), -float('inf')], 0).fillna(0)
combined_df
```

```python
import matplotlib.pyplot as plt

# 1. 폰트 설정 (한글 깨짐 방지)
plt.rcParams['font.family'] = 'Malgun Gothic' # 윈도우 기준
plt.rcParams['axes.unicode_minus'] = False

# 2. 모든 영양소에 대해 막대그래프 그리기
plt.figure(figsize=(15, 8))

# x축은 영양소 인덱스, y축은 계산하신 '섭취비율(%)'
# 만약 '섭취비율(%)' 열이 아직 없다면 아래 주석을 해제하고 실행하세요.
# df_final['섭취비율(%)'] = (df_final['실제섭취량'] / df_final.iloc[:, 1]) * 100

bars = plt.bar(combined_df.index, combined_df['섭취비율(%)'], color='skyblue', edgecolor='navy', alpha=0.7)

# 3. 100% 위치에 빨간색 기준선 그리기
plt.axhline(y=100, color='red', linestyle='--', linewidth=2, label='권장 섭취량 기준 (100%)')

# 4. 그래프 꾸미기
plt.title('영양소별 권장량 대비 실제 섭취 비율 - 여성 (%)', fontsize=18, pad=20)
plt.ylabel('섭취 비율 (%)', fontsize=12)
plt.xlabel('영양소', fontsize=12)
plt.xticks(rotation=45, ha='right') # 영양소 이름이 길 경우를 대비해 회전
plt.grid(axis='y', linestyle=':', alpha=0.6)
plt.legend()

# 5. 막대 위에 수치 표시 (가독성 향상)
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height + 1,
             f'{height:.1f}%', ha='center', va='bottom', fontsize=9)

plt.tight_layout()
plt.show()
```
