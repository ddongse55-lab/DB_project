```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 경고 무시 설정
import warnings
warnings.filterwarnings(action='ignore')

# 한글 깨짐 방지
plt.rcParams['font.family'] = 'Malgun Gothic'
# 마이너스 부호 깨짐 방지
plt.rcParams['axes.unicode_minus'] = False
```

## 남성 Total

```python
df01 = pd.read_csv('./남성 영양소별 권장 섭취량(연령별)_전체.csv')
df01
```

```python
df01 = df01[['에너지(kcal)', '단백질(g)', '탄수화물(g)',
             '칼슘(mg)', '인(mg)', '철(mg)', '나트륨(mg)', '칼륨(mg)',
            '비타민A(μgRAE)', '티아민(mg)', '리보플라빈(mg)', '나이아신(mg)',
            '비타민C(mg)', '엽산(μgRAE)']]
df01 = df01.dropna()
df01
```

```python
df = df01.iloc[8]
df
```

```python
df01_usage = pd.read_csv('./영양소별 섭취량 - 성별(남).csv')
df01_usage
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. 한글 폰트 설정 (윈도우 기준)
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False

# 2. 데이터 준비 및 전처리
# df01_usage의 '영양소' 컬럼을 인덱스로 설정
df01_usage_indexed = df01_usage.set_index('영양소')

# 3. 데이터 매칭을 위한 인덱스 통일
# df(권장량)와 df01_usage(실제섭취량)의 영양소 순서가 동일하므로 df의 인덱스를 복사하여 사용합니다.
# 이는 단위나 띄어쓰기 차이로 인한 매칭 오류를 방지합니다.
df01_usage_indexed.index = df.index

# 4. 백분율 계산 (실제 섭취량 / 권장 섭취량 * 100)
# '평균' 컬럼의 수치를 df의 수치로 나눕니다.
ratio = (df01_usage_indexed['평균'] / df) * 100

# 5. 막대그래프 시각화
plt.figure(figsize=(12, 8))
bars = plt.bar(ratio.index, ratio.values, color='skyblue', edgecolor='black', alpha=0.8)

# 6. 100% 기준선 추가 (사용자 요청: 빨간선)
plt.axhline(y=100, color='red', linestyle='--', linewidth=2, label='권장량 기준 (100%)')

# 7. 그래프 세부 설정
plt.title('영양소별 권장량 대비 실제 섭취 비율 - 남성 (%)', fontsize=16, pad=20)
plt.ylabel('섭취 비율 (%)', fontsize=12)
plt.xlabel('영양소', fontsize=12)
plt.xticks(rotation=45, ha='right')  # 라벨 겹침 방지
plt.grid(axis='y', linestyle=':', alpha=0.7)
plt.legend()

# 막대 위에 구체적인 수치 표시
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height + 1,
             f'{height:.1f}%', ha='center', va='bottom', fontsize=9)

plt.tight_layout()
plt.show()
```
