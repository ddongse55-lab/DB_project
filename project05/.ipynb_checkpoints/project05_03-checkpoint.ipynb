```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 경고 무시 설정
import warnings
warnings.filterwarnings(action='ignore')

# 한글 깨짐 방지
plt.rcParams['font.family'] = 'Malgun Gothic'
# 마이너스 부호 깨짐 방지
plt.rcParams['axes.unicode_minus'] = False
```

## 부산 영도구 노인요양시설

```python
df00 = pd.read_csv('./영도구_장기요양기관내역.csv')
df00
```

```python
df01 = df00['급여종류'].value_counts()
df01
```

```python
df01.plot(kind='bar', color='skyblue', figsize=(10, 6))
plt.title('영도구 급여종류별 장기요양기관 수')
plt.xlabel('급여종류')
plt.ylabel('기관 수')
plt.xticks(rotation=45)
plt.show()
```
